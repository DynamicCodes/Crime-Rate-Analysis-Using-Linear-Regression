## CHECKING THE OUTLIERS IN THE TARGET VARIABLE ##

```python
# In our case the number of victims of all age groups is the target variable which our model has to predict
```

```python
# Any data point below lower_limit and any data point above upper_limit will be treated as an outlier.
def treat_outlier(value):
    if value<lower_limit:
        return lower_limit
    elif value>upper_limit:
        return upper_limit
    else:
        return value
```
